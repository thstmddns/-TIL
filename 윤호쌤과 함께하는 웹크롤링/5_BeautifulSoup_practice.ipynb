```python
import requests as req
```

```python
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.cadillac.co.kr/crossovers-suvs/escalade'
res = req.get(url)
res.text
```

### Beautifulsoup 객체화

```python
bs_cadi = bs(res.text, 'lxml')
bs_cadi
```

```python
car_name = bs_cadi.select('h4.q-headline.none-margin')
car_name[1].text
```

```python
# replace() 함수를 사용, strip()함수를 사용할 겁니다.

# for loop

for name in car_name:
    # 공백을 제거한 것처럼 보이나 실제로 공백이 제거되지 않았음
    # 그래서 replace만 진행이 된다면, 개행문자가 빈문자로 대체될 뿐, 아직 공백은 사라지는 것이 아님
    # 그래서 for loop을 돌리는 김에, strip()함수까지 써서 한큐에 보내야 함
    car_name_strip = name.text.replace('|n', '').replace('|t', '').strip()
    print(car_name_strip)
    
```

### 옵션추출

```python
car_option = bs_cadi.select('#q-main-content > adv-grid:nth-child(22) > adv-col:nth-child(2) > div > adv-grid > adv-col.col-sm-12.col-xl-12.col-xl-pad-lt-3 > div > div.none-margin > div')
car_option[0].text
```

```python
for option in car_option:
    option_strip = option.text.replace('|n', '').replace('|t', '').strip()
    print(option_strip)
```

```python
features = bs_cadi.select('div.none-margin li')
for i in features:
    print(i.text)
```

```python
# 옵션'들'을 추출
    # 1. 빈 리스트 (빈 그릇)
suv_features = []
for feature in features:
    # for loop을 사용 -> feature 리스트의 각 요소를feature로 선언한 변수에 순차적으로 넣어주면서 
    suv_features.append(feature.text.strip())
suv_features
```

```python
import pandas as pd
```

```python
car = pd.DataFrame({'옵션들' : suv_features
})
car.index.name = 'SPORT PLATINUM'
```

```python
car
```

```python
car.to_csv('data/pls_giv_me.csv', encoding='euc-kr')
```
