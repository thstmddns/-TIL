### Melon Chart top 100수집 [타이틀|아티스트]

- 타이틀과 아티스트 명을 beautifulsoup 사용해서 수집
- for loop 적용
- 마지막으로 데이터를 Pandas Library의 Dataframe에 집어넣기
- 데이터 저장 파일을 내보내기

```python
import requests as req
from bs4 import BeautifulSoup as bs

import pandas as pd
```

```python
url = 'https://www.melon.com/chart/'
req.get(url)
```

### BeautifulSoup 객체화


```python
soup = bs(melon_chart.text, 'lxml')
soup
```

### 수집한 타이틀, 아티스트 데이터를 for loop 통해 리스트에 집어넣자

```python
# 자료형을 확인해보자
type(titles)

# bs4.element.ResultSet이네?
    # 리스트와 매우 유사하지만 엄연히 다른 객체
    # but 우리는 python기반 pandas library를 사용해야한다.
    # 그러므로 빈 리스트 2개를 선언해서  for loop으로 순차적으로 담자
```

```python
# 빈 리스트를 선언
titles_list = []
artist_list = []
for i in range(len(titles)):
    titles_list.append(titles[i].text)
    artist_list.append(artist[i].text)
#     print(artist[i].text)
    
    print(titles[i].text, artist[i].text)
```

DataFrame

```python
# 1. Dictionary 형태로 만들어 줘야합니다.
    # Dictionary로 우선 행과선언해 줘야하는 이유:
    # 1. 각각의 행과 열에 대한 데이터를 쉽게 입력할 수 있다.
    # 2. frame 내 데이터 타입을 일치시키기가 쉽게 구성할 수 있다.
    # 3. dictionary로 먼저 저장을 해주면, 
    # 데이터 구조가 보존이 됩니다.
    # Pandas DataFrame으로 변경할 때, 일관성이 유지됩니다.
```

```python
melon_dic = {'타이틀' : titles_list,
            '아티스트' : artist_list}
melon_dic
```

```python
melon_df = pd.DataFrame(melon_dic)
melon_df
```

```python
# CSV파일로 내보내기
    # 항상 파일로 내보내는 습관이 필요하다.
    # 나중에 다시 다 돌려서, 다시 뽑아야 하는 불상사가 발생
melon_df.to_csv('C:/Users/user04/윤호쌤과 함께하는 웹크롤링/data/melon_chart.csv', encoding ='euc-kr')
```
