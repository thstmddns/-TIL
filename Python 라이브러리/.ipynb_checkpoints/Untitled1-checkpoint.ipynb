```python
# 3.)
import matplotlib.pyplot as plt

X = [2, 3, 4, 5, 6, 7, 8]
Y1 = [5, 7, 3, 9, 1, 8, 6]
Y2 = [9, 5, 7, 2, 3, 7, 1]

from matplotlib import rc
rc('font', family = 'batang')

pl1 = plt.plot(X, Y1, c = 'r', label = 'Y1')
pl2 = plt.plot(X, Y2, c = 'b', label = 'Y2')
plt.grid()
plt.title('<실행결과>')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc = 'upper right', prop = {'size' : 10})
plt.show(pl1,pl2)
```

```python
# 4.)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/Traffic_Accident_2017.csv', encoding = 'euc-kr')
```

```python
from matplotlib import rc
rc('font', family = 'batang')
```

```python
ac = data[(data['사고유형_대분류']=='차대사람') & (data['주야']=='야간')]
```

```python
ac2 = ac['요일'].value_counts()
ac2 = ac2.reindex(['월', '화', '수', '목', '금', '토', '일'])
```

```python
x = np.arange(7)
y = ac2
```

```python
plt.bar(x,y)
plt.xticks(x, y.index)
plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.ylim(0,200)
plt.show()
```
