### Matplotlib
- 데이터 분석 결과를 그래프나 차트로 시각화 해주기 위한 라이브러리
- 분석 결과를 수치만으로 설명하는 것보다 시각적인 도구들을 활용하면 훨씬 더 직관적으로 이해가 가능하고 타인에게 전달하기도 용이함

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 플롯 차트(plot chart) : 꺾은선 그래프

```python
# x값 없이 선 그래프 그려보기
y = [2,4,6]

# plot : 꺾은선 그래프 출력 함수
plt.plot(y)
# show : 차트를 화면에 표시해주는 함수(쥬피터 노트북에서는 안써도 인식됨)
plt.show()
```

```python
# x, y값 지정하여 선 그래프 그려보기
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y)
plt.show()
```

#### 차트 style 설정

#### linestyle

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.')
plt.show()
```

#### Line width

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

# lw는 숫자로 설정 가능
plt.plot(x, y, ls='-.', lw=4)
plt.show()
```

#### Marker

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=3, marker='D')
plt.show()
```

#### Line color

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=3, marker='D', c='c')
plt.show()
```

#### 차트 옵션


#### x축, y축 범위 설정

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x, y, ls='--', lw=4, marker='o', ms=15, mec='g', mew=5, mfc='r')
# x축 표시 범위 설정
plt.xlim(-2, 10)
# y축 표시 범위 설정
plt.ylim(-5, 12)
plt.show()
```

#### 틱(tick) 설정
- 플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 함
- 이 틱 위에 써진 숫자 혹은 문자는 틱 라벨(tick label)이라고 함
- 틱 위치나 라벨은 matplotlib에서 자동으로 정해주지만 수동 설정을 하고 싶다면 xticks, yticks 명령을 사용

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x, y, ls='--', lw=4, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])
plt.show()
```

#### 그리드(grid)
- 그림의 틱 위치를 잘 보여주기 위해 중간에 그리드 선(grid line)을 그어주는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x, y, ls='--', lw=4, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])
plt.grid()
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x, y, ls='--', lw=4, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.grid()
plt.show()

# tick이 있는 지점에만 grid가 생성됨!
```

#### x축, y축 제목 및 그래프 제목

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x, y, ls='--', lw=4, marker='o', ms=15, mec='g', mew=5, mfc='r')
# x축 제목 설정
plt.xlabel("X axis")
# y축 제목 설정
plt.ylabel("Y axis")
# 그래프 전체 제목 설정
plt.title("X-Y graph")

plt.grid()
plt.show()
```

#### 범례 설정 + 여러개 플롯 한꺼번에 그리기!

```python
x = np.arange(4)

y1 = [1,2,3,4]
y2 = [2,4,6,8]
y3 = [3,6,9,12]

# 범례 설정을 위해 각 plot마다 라벨 설정
plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y2")
plt.plot(x, y3, label="y3")
# 범례 설정
# loc : 범례의 위치 표시(upper, center, lower) + (left, center, right)
# prop : 범례 크기 조절
plt.legend(loc='upper right', prop={'size':7})
plt.show()
```

### 바 차트(bar chart) : 막대 그래프

```python
x = np.arange(6)
y = [80,85,70,60,50,90]
```

```python
# bar : 바 차트 설정
plt.bar(x, y)
plt.show()
```

```python
# 막대의 색상 변경
plt.bar(x, y, color='orange')
plt.show()
```

```python
# 원하는 막대의 색상 변경
plt.bar(x, y, color=['orange', 'cyan', 'pink'])
plt.show()
```

```python
# edgecolor : 막대 테두리 색상 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue')
plt.show()
```

```python
# linewidth : 막대 테두리 굵기 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue', linewidth=4)
plt.show()
```

```python
# alpha : 투명도 설정(0~1사이의 실수값, 1이면 완전 불투명, 0이면 완전 투명)
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue', linewidth=4, alpha=0.5)
plt.show()
```

```python
# width : 막대 너비 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='blue',
        linewidth=4, alpha=0.5, width=0.5)
plt.show()
```

#### 2) 요일별 사고 건수 시각화
- 요일별 사고 건수 확인
- 요일 순서대로 설정 후 변수에 저장해서 시각화에 활용

```python
# 요일별 사고 건수 확인
days = data['요일'].value_counts()  # 내림차순이 디폴트
days
```

```python
# days.index = ['월', '화', '수', '목', '금', '토', '일']
# days

# # 인덱스 자체만 바뀜
```

```python
# reindex : 인덱스와 그에 대응하는 값의 순서를 같이 변경
days = days.reindex(['월', '화', '수', '목', '금', '토', '일'])
days
```

#### 3) 요일별 전국 교통사고 발생건수 시각화

#### 연습문제
### 사고유형이 차대차 사건 중에 죽거나 다친사람이 많은 발생지시도를 알아보고 시각화해보자!

- 중요 키워드 3개 ? : 차대차, 죽거나 다친사람, 발생지시도

```python
data.columns
```

```python
data['사고유형_대분류'].value_counts()
```

```python
data['사상자수'].value_counts()
```

```python
data['발생지시도'].value_counts()
```

```python
# 사고유형_대분류에서 차대차 사건만 불리언 인덱싱
ac = data[data['사고유형_대분류']=='차대차']
ac
```

```python
ac.shape
# 사고유형이 차대차인 사건은 총 1642건
```

```python
ac[['발생지시도', '사상자수']]
```

```python
# 1642개의 사건 중에서 각 시도의 빈도수 출력
ac['발생지시도'].value_counts()
```

```python
# 각 발생지시도 별로 사상자수의 합계 확인하기
# 그룹 별(발생지시도)로 묶어서 집계(사상자수의 합)를 출력
```

```python
ac2 = ac[['발생지시도','사상자수']].groupby('발생지시도').sum()
ac2
```

```python
# 발생지시도의 개수
len(ac2)
```

```python
ac2.index
```

#### 실습 예제
위 데이터를 활용하여 아래와 같은 결과 그래프를 그려보세요!
- 1.bar차트의 x, y축에 넣을 값 설정
  - x에는 발생지시도, y에는 시도별 사상자수의 합이 들어가야하며 내림차순 정렬
- 2.그래프의 세부사항들 설정


```python
# x,y축 값 설정하기
x = np.arange(len(ac2.index))  # 17개의 시도
y = ac2['사상자수'].sort_values(ascending=False)
```

```python
# ac2['사상자수'].sort_values(ascending=False)
```

```python
# 그래프 그리기
plt.figure(figsize=(15,7))     # 그래프 전체 가로세로 사이즈
plt.bar(x, y)
plt.xticks(x, y.index)
plt.xlabel('지역')
plt.ylabel('사상자 수')
plt.title('차vs차 교통사고의 사상자 수')
plt.show()
```
