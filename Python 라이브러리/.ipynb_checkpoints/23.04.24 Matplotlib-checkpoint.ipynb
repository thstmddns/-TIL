#### Matplotlib
- 데이터 분석 결과를 그래프나 차트로 시각화 해주기 위한 라이브러리
- 분석 결과를 수치만으로 설명하는 것보다 시각적인 도구들을 활용하면 훨씬 더 직관적으로 이해가 가능하고 타인에게 전달하기도 용이함

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### 플롯 차트(plot chart):꺽은선 그래프

```python
#x값 없이 선 그래프 그려보기
y = [2, 4, 6]
# plot : 꺽은선 그래프 출력 함수
plt.plot(y)
# show : 차트를 화면에 표시해주는 함수(주피터 노트북에서는 안써도 인식됨)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]
plt.plot(x, y)
plt.show()
```

#### 차트 스타일 설정

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]
plt.plot(x, y, ls =':')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]
plt.plot(x, y, ls='-.', lw= 4)
plt.show()
```

#### Marker


```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]
plt.plot(x, y, ls='-.', lw= 4, marker='d')
plt.show()
```

#### Line color


```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]
plt.plot(x, y, ls='-.', lw= 4, c = 'm')
plt.show()
```

#### 실습예제
- 아래와 같이 그래프를 그려보아요


```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls = '--', lw = 1, marker = 'p', c = 'g')
plt.show()
```

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls = ':', lw = 1, c = 'r', marker = 'D')
plt.show()
```
