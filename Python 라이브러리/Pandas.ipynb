#### apply
- 행 또는 열에 복잡한 처리 및 연산을 한 번에 진행해주는 함수
- apply함수 사용을 위해서는 사전에 사용자 정의 함수가 선언되어 있어야 함(우리가 만든 사용자 정의 함수의 로직에 따라서 적용시켜주기 때문)

```python
#drop에서만 axis = 0과 1이 다름을 주의!!
score_data.drop(['총합', '평균'], axis = 1, inplace = True)
score_data
```

```python
#최대 값, 최솟 값의차이를 계산해주는 사용자 정의 함수 선언
def cal(x):
    return x.max() - x.min()
```

```python
# DF 내 전체 데이터에서 각 과목의 최대 점수와 최소 점수의 차를 구하기
# 사용자정의 함수인 cal에 들어가는 매개변수 x는 apply함수의 바로 앞 변수인 score_data가 들어가게 됨

score_data.apply(cal, axis = 1)
```

#### 정리 
- apply 함수에는 반드시 사용자 정의 함수가 포함되어야 함
- 사용자 정의 함수에 입력되는 매개변수는 apply함수 바로 앞에 오는 변수가 돌아오게 됨

### Pandas 데이터 분석 실습 예제!
2015 ~ 2017년 광주광역시 범죄현황 데이터를 이용해 전년 대비 지역별 범죄 증감률을 구해보자!
- 증감율 = (금년-작년)/작년 * 100
    - ex)
- 아래 화면과 똑같은 결과를 출력시켜보자


```python
data_2015 = pd.read_csv('data/2015.csv', encoding='euc-kr', index_col='관서명')
data_2015
```

```python
data_2016 = pd.read_csv('data/2016.csv', encoding='euc-kr', index_col='관서명')
data_2016
```

```python
data_2017 = pd.read_csv('data/2017.csv', encoding='euc-kr', index_col='관서명')
data_2017
```

```python
data_2017.drop('광주지방경찰청', inplace = True)
data_2017
```

```python
# 총계 구학기
da2015 = data_2015.loc['광주지방경찰청계':'광주광산경찰서':6, '살인':].sum(axis = 1)
```

```python
da2016 = data_2016.loc['광주지방경찰청계':'광주광산경찰서':6, '살인':].sum(axis = 1)
```

```python
da2017 = data_2017.loc['광주지방경찰청계':'광주광산경찰서':6, '살인':].sum(axis = 1)
```

```python
da2017
```

```python
das2016 = (da2016 - da2015)/da2015 * 100
```

```python
das2017 = (da2017 - da2016)/da2016 * 100
```

```python
result = pd.DataFrame
result
```
